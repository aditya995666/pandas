# join methon

```python
import pandas as pd
```

```python
df1=pd.DataFrame({'A':['A0','A1','A2','A3'],'B':['B0','B1','B2','B3']},index=[1,2,3,44])
df1
```

```python
df2=pd.DataFrame({'A':['A4','A5'],'B':['B6','B7']},index=[3,44])
df2
```

```python
df1.join(df2,lsuffix=1)
```

```python
df1.join(df2,lsuffix=1,how="inner")
```

```python
df1.join(df2,lsuffix=1,how="outer")
```

```python
df1.join(df2,lsuffix=1,how="left")
```

```python
df1.join(df2,lsuffix=1,how="right")
```
