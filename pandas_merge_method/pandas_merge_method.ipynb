### merge method one and more dataframe

```python
import pandas as pd
```

```python
df1=pd.DataFrame({"ID":[1,2,3,4,5],
                 "class":[9,10,11,12,13]})
df1
```

```python
df2=pd.DataFrame({"ID":[1,2,3,4,6],
                  "name":["A","B","C","D","E"]})
df2
```

```python
pd.merge(df1,df2)
```

```python
pd.merge(df2,df1)
```

```python
pd.merge(df1,df2,on="ID")
```

```python
pd.merge(df2,df1,on="ID")
```

```python
pd.merge(df1,df2,on="ID",how="inner")
```

```python
pd.merge(df1,df2,on="ID",how="outer")
```

```python
pd.merge(df1,df2,on="ID",how="left")
```

```python
pd.merge(df2,df1,on="ID",how="left")
```

```python
pd.merge(df2,df1,on="ID",how="right")
```

```python
df1=pd.DataFrame({"ID":[1,2,3,4,5],
                 "class":[9,10,11,12,13]})
df1
```

```python
df2=pd.DataFrame({"ID":[7,8,9,0,1],
                  "name":["A","B","C","D","E"]})
df2
```

```python
pd.merge(df1,df2,left_index=True,right_index=True)
```
