```python
import pandas as pd
```

```python
df=pd.read_excel("/content/Book1.xlsx")
df.head()
```

```python
# use pivate_table
df.pivot_table(index="gegre")
```

```python
# use pivate_table
df.pivot_table(index="gegre",columns="year")
```

```python
df.pivot_table(index="gegre",columns="year",aggfunc="sum")
```

```python
df.pivot_table(index="gegre",columns="year",aggfunc="sum",fill_value=0)
```

```python
df.pivot_table(index="gegre",columns="year",aggfunc="sum",fill_value=0,margins=True)
```
