# fillna methon

```python
import pandas as pd
```

```python
df=pd.read_csv("/content/dataset_fillna.csv")
df
```

```python
df.fillna(4)
```

```python
# fillna value use of dict
df.fillna({"Gender":"root","ParentalSupport":0})
```

```python
df.fillna(0,axis=0)
```

```python
df.fillna(method="ffill")
```

```python
df.fillna(method="bfill")
```

```python
df.fillna(axis=1,method="bfill")
```

```python
df.fillna(0,limit=1)
```

```python
df.fillna(method="ffill",limit=1)
```

```python
df.fillna(2)
```

```python
df.fillna(method="ffill",limit=2)
```
