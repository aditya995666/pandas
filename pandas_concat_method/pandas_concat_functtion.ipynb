# series , dataframe and panel object ko concate()

```python
import pandas as pd
```

```python
# create a series_1
sr=pd.Series([0,1,2,3,4])
sr
```

```python
# create a series_2
sr1=pd.Series([5,6,7,8,9])
sr1
```

```python
pd.concat([sr,sr1],axis=1)
```

```python
# create a dataframe_1
df_1=pd.DataFrame({"ID":[0,1,2,3],
                 "CLASS":[5,6,7,8],
                 "SEC":["A","B","C","D"]})
df_1
```

```python
# create a dataframe_2
df_2=pd.DataFrame({"ID":[4,5,6,7],
                 "CLASS":[9,10,11,12],
                 "SEC":["E","F","G","H"]})
df_2
```

```python
pd.concat([df_1,df_2],axis=0)
```

```python
pd.concat([df_1,df_2],axis=0,ignore_index=True)
```
